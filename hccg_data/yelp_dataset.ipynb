```python
import requests
import json
```

```python
def pull_reviews_per_location(loc_url):
    raw = requests.get(loc_url).text
    print(raw)
    if not is_cake(raw):
        return False
    return raw
```

```python
def is_cake(test_html):
    trimmed_html = test_html[test_html.find('<meta property="og:title" content="'):]
    trimmed_html = trimmed_html[len('<meta property="og:title" content="'):trimmed_html.find(">\n")]
    if "The Cheesecake Factory" in trimmed_html:
        return True
    return False
```

```python
review_bodies = []
def extract_reviews(raw_html):
    trimmed_html = raw_html
    while raw_html.find('<div class=" review') != -1:
        print("Found one!")
        trimmed_html = trimmed_html[trimmed_html.find('<div class=" review__')+len('<div class=" review__'):]
        review_body = trimmed_html[trimmed_html.find('<span class=" raw__'):trimmed_html.find("</span>")]
        review_bodies.append(review_body)
```

```python
test_page = pull_reviews_per_location("https://www.yelp.com/biz/the-cheesecake-factory-cambridge?osq=The+Cheesecake+Factory")
```
