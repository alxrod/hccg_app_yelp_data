```python
import requests
import json
```

```python
link_file = open("links.txt", "r")
links = link_file.read().split("\n")
```

```python
def interp_json(url):
    string_json="CTYPE HTML>"
    while "CTYPE HTML>" in string_json:
        print('Failed load....')
        
        raw = requests.get(url).text

        string_json = raw[raw.find('><!--{"gaConfig": {"metrics":')+len('><!--'):]
        string_json = string_json[0:string_json.find("-->")]

#     print(string_json)
    json_content = json.loads(string_json)
    return json_content
```

```python
def handle_location(url):
    location_data = {}
    
    json_content = interp_json(url)
    if json_content["bizDetailsPageProps"]["businessName"] != "The Cheesecake Factory":
        return
    
    location_data["address"] = json_content["bizDetailsPageProps"]["mapBoxProps"]["addressProps"]["addressLines"]
    
    location_data["monthly_reviews"] = {}
    review_years = json_content["bizDetailsPageProps"]["ratingDetailsProps"]["monthlyRatingsByYear"].keys()
    for year in review_years:
        location_data["monthly_reviews"][int(year)] = json_content["bizDetailsPageProps"]["ratingDetailsProps"]["monthlyRatingsByYear"][year]
    
    location_data["reviews"] = []
    
    page_start = json_content["bizDetailsPageProps"]["reviewFeedQueryProps"]["pagination"]["startResult"]
    perPage = json_content["bizDetailsPageProps"]["reviewFeedQueryProps"]["pagination"]["resultsPerPage"]
    totalResults = json_content["bizDetailsPageProps"]["reviewFeedQueryProps"]["pagination"]["totalResults"]
    
    while page_start + perPage <= totalResults: 
        print("Processing a new page of reviews...")
        json_content = interp_json(url+"&start="+str(page_start+perPage))
        page_start = json_content["bizDetailsPageProps"]["reviewFeedQueryProps"]["pagination"]["startResult"]
        perPage = json_content["bizDetailsPageProps"]["reviewFeedQueryProps"]["pagination"]["resultsPerPage"]
    
        for temp_rev in json_content["bizDetailsPageProps"]["reviewFeedQueryProps"]["reviews"]:
            review = {}
            review["text"] = temp_rev["comment"]["text"]
            review["rating"] = temp_rev["rating"]
            review["date"] = temp_rev["localizedDate"]
            location_data["reviews"].append(review)
    
    return location_data
```

```python
links = links[49:]
for link in links:
    print("\n\n\n\n\nSTARTING NEW LOCATION")
    ld = handle_location(link)
    name = ld["address"][-1].split(" ")[-1]
    with open("yelp_data/"+name+".json", 'w+') as outfile:
        json.dump(ld, outfile)
```
